```python
import os
import pandas as pd
import numpy as np

# Change to Downloads folder
os.chdir(os.path.expanduser('~/Downloads'))

file_path = 'merged file (1).xlsx'

# Load and process data
stock_data_raw = pd.read_excel(file_path, sheet_name='close', header=None)

# Extract tickers
tickers = stock_data_raw.iloc[0, 1:].tolist()

# Build stock price DataFrame
stock_data = pd.DataFrame()
stock_data['Date'] = stock_data_raw.iloc[2:, 0]

for i, ticker in enumerate(tickers):
    stock_data[ticker] = stock_data_raw.iloc[2:, i + 1]

# Convert dates and clean
stock_data['Date'] = pd.to_datetime(stock_data['Date'])
stock_data.set_index('Date', inplace=True)
stock_data = stock_data.apply(pd.to_numeric, errors='coerce').dropna(how='all')

# Calculate daily returns
returns = stock_data.pct_change().dropna()

# Calculate 30-day rolling correlation matrices
rolling_corrs = (
    returns.rolling(window=30)
    .corr()
    .dropna()
)

# Save the final rolling correlation matrices
rolling_corrs.to_csv('rolling_corr_matrix_30D.csv')

print("30-day rolling correlation matrix calculated and saved as 'rolling_corr_matrix_30D.csv'.")
```
